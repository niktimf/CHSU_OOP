Разберите, реализуйте в системе программирования и протестируйте программу, соответствующую примеру лекции: Пример 2 (слайд 9).
```python
from typing import List
import random

class Point:
    """ Класс, описывающий точку на плоскости. """
    def __init__(self, x=0, y=0):
        """ Инициализация точки с координатами x и y. """
        self.x = x
        self.y = y

    def __str__(self):
        """ Вывод координат точки в формате (x, y). """
        return f"({self.x}, {self.y})"

    def __repr__(self):
        """Представление точки для вывода в списках и других структурах. """
        return self.__str__()


class Circle(Point):
    """ Класс, описывающий круг на плоскости. """
    def __init__(self, x=0, y=0, r=0):
        """ Инициализация круга с координатами центра x, y и радиусом r. """
        super().__init__(x, y)
        self.r = r

    def distance(self, point):
        """ Расчет расстояния от центра круга до указанной точки. """
        return ((point.x - self.x) ** 2 + (point.y - self.y) ** 2) ** 0.5

    def __str__(self):
        """ Вывод информации о круге. """
        return f"Круг с центром в точке: ({self.x}, {self.y}) и радиусом: {self.r}"



def create_circle_object():
    """ Создание объекта Circle на основе пользовательского ввода. """
    x = int(input("Введите координату x: "))
    y = int(input("Введите координату y: "))
    r = int(input("Введите радиус r: "))
    return Circle(x, y, r)


def generate_random_coords(range_min=0, range_max=20) -> Point:
    """ Генерация случайных координат для точки. """
    x = random.randint(range_min, range_max)
    y = random.randint(range_min, range_max)
    return Point(x, y)


def generate_random_circle_coords() -> List[Point]:
    """Генерация списка из n точек со случайными координатами."""
    n = int(input("Введите количество точек: "))
    return list(map(generate_random_coords, range(n)))


def farthest_point_from_circle(circle, points_list):
    """Находит точку, которая наиболее удалена от центра указанного круга."""
    # Вычисляем расстояния от центра круга до каждой точки
    distances = [circle.distance(point) for point in points_list]
    # Находим индекс точки с наибольшим расстоянием
    max_index = distances.index(max(distances))
    return max_index, points_list[max_index]



circle = create_circle_object()
random_points = generate_random_circle_coords()

print(circle)
print(f"Все точки: {random_points}")
print(f"Точка расположенная дальше всех от центра круга и ее координаты: {farthest_point_from_circle(circle, random_points)}")
```

Создать класс треугольник, члены класса – координаты 3-х точек. Предусмотреть в классе методы проверки существования треугольника, вычисления указанных параметров и вывода всех сведений о фигуре. Параметры: длины сторон, углы, периметр, площадь. Создать производный класс – прямоугольный треугольник, предусмотреть в классе проверку, является ли треугольник прямоугольным. Написать программу, демонстрирующую работу с классом: дано N треугольников и M прямоугольных треугольников, найти среднюю площадь и минимальный периметр для N треугольников и прямоугольный треугольник с наибольшей гипотенузой.

```python
import math
import pandas as pd


class Triangle:
    """ Класс, описывающий треугольник. """
    def __init__(self, A, B, C):
        self.A = A
        self.B = B
        self.C = C


    def side_length(self, point1, point2):
        """Вычисление длины стороны между двумя точками."""
        side_length = math.sqrt((point2.x - point1.x) ** 2 + (point2.y - point1.y) ** 2)
        return side_length


    def sides_length(self):
        """Вычисление всех длин сторон в треугольнике."""
        sides = [self.side_length(p1, p2) for p1, p2 in zip([self.A, self.B, self.C], [self.B, self.C, self.A])]
        return sides


    def exists(self):
        """Проверка существования треугольника."""
        # Вычисляем длины сторон
        #print(sides)
        # Проверка, что сумма длин любых двух сторон больше третьей
        if not all(a < sum(self.sides_length()) - a for a in self.sides_length()):
            return False

        # Проверка, что вершины не лежат на одной прямой
        determinant = self.A.x * (self.B.y - self.C.y) + self.B.x * (self.C.y - self.A.y) + self.C.x * (self.A.y - self.B.y)
        return determinant != 0


    def angles(self):
        """Возвращает углы треугольника в градусах."""
        alpha = math.degrees(math.acos((self.sides_length()[1] ** 2 + self.sides_length()[2] ** 2 - self.sides_length()[0] ** 2) / (2 * self.sides_length()[1] * self.sides_length()[2])))
        beta = math.degrees(math.acos((self.sides_length()[0] ** 2 + self.sides_length()[2] ** 2 - self.sides_length()[1] ** 2) / (2 * self.sides_length()[0] * self.sides_length()[2])))
        gamma = math.degrees(math.acos((self.sides_length()[0] ** 2 + self.sides_length()[1] ** 2 - self.sides_length()[2] ** 2) / (2 * self.sides_length()[0] * self.sides_length()[1])))
        return alpha, beta, gamma


    def perimeter(self):
        """Вычисление периметра треугольника."""
        if not self.exists():
            return 0
        perimeter = sum(self.sides_length())
        return perimeter


    def area(self):
        """Вычисление площади треугольника."""
        if not self.exists():
            return 0
        s = sum(self.sides_length()) / 2
        area = math.sqrt(s * (s - self.sides_length()[0]) * (s - self.sides_length()[1]) * (s - self.sides_length()[2]))
        return area


    def __str__(self):
        return f"{self.A}, {self.B} и {self.C}"




class RightTriangle(Triangle):
    """ Класс, описывающий прямоугольный треугольник(наследуется от треугольника). """
    def hypotenuse(self):
        """ Вычисление гипотенузы прямоугольного треугольника """
        hypotenuse = max(self.sides_length())
        return hypotenuse

    def is_right(self):
        """Проверка, является ли треугольник прямоугольным."""
        other_sides = list(filter(lambda x: x != self.hypotenuse(), self.sides_length()))
        # Проверка теоремы Пифагора
        return self.hypotenuse() ** 2 == sum([x ** 2 for x in other_sides])



def generate_third_point(A, B, range_min, range_max):
    """Генерация третьей точки, которая не лежит на прямой AB."""
    # Если точки A и B имеют одинаковые x-координаты, прямая AB вертикальна.
    if A.x == B.x:
        # В этом случае просто выберем случайное значение x, отличное от A.x и B.x
        x = random.choice([i for i in range(range_min, range_max + 1) if i not in [A.x, B.x]])
        y = random.randint(range_min, range_max)
    else:
        # Найдем уравнение прямой AB в форме y = mx + b
        m = (B.y - A.y) / (B.x - A.x)
        b = A.y - m * A.x
        # Выберем случайное значение x и вычислим y на основе уравнения прямой.
        # Затем добавим или вычтем некоторое случайное значение, чтобы точка C не лежала на прямой AB.
        x = random.randint(range_min, range_max)
        y = m * x + b + random.choice([-1, 1]) * random.randint(1, range_max - range_min)
        # Удостоверимся, что y находится в допустимом диапазоне.
        y = max(min(y, range_max), range_min)

    return Point(x, y)


def generate_pythagorean_triple(max_value=10):
    """Генерация пифагоровой тройки."""
    m = random.randint(2, max_value)
    # m должно быть больше чем n
    n = random.randint(1, m - 1)

    a = m ** 2 - n ** 2
    b = 2 * m * n
    c = m ** 2 + n ** 2
    return a, b, c


def generate_random_triangle(range_min=0, range_max=10):
    """Генерация треугольника с двумя случайными точками и одной вычисленной."""
    A = generate_random_coords(range_min, range_max)
    B = generate_random_coords(range_min, range_max)
    C = generate_third_point(A, B, range_min, range_max)
    triangle = Triangle(A, B, C)

    # Если по каким-то причинам получившийся треугольник все равно не существует, генерируем его заново.
    while not triangle.exists():
        A = generate_random_coords(range_min, range_max)
        B = generate_random_coords(range_min, range_max)
        C = generate_third_point(A, B, range_min, range_max)
        triangle = Triangle(A, B, C)
    return triangle



def generate_random_right_triangle(range_min=0, range_max=10):
    """Генерация прямоугольного треугольника с целыми длинами сторон."""

    # Генерация двух случайных точек A и B
    A = generate_random_coords(range_min, range_max)
    B = generate_random_coords(range_min, range_max)

    # Генерация пифагоровой тройки
    a, b, c = generate_pythagorean_triple(max_value=range_max)

    # Выбираем, какой катет будет горизонтальным, а какой вертикальным
    if random.choice([True, False]):
        a, b = b, a

    # Вычисляем координаты третьей точки C
    C = Point(A.x + a, A.y + b)

    return RightTriangle(A, B, C)


def input_triangles():
    """Ввод данных о треугольниках и генерация списков треугольников."""
    N = int(input("Введите количество треугольников N: "))
    triangles_obj = [generate_random_triangle() for _ in range(N)]
    triangles_coords = [(triangle.A, triangle.B, triangle.C) for triangle in triangles_obj]
    return triangles_obj, triangles_coords


def input_right_triangles():
    """Ввод данных о прямоугольных треугольниках и генерация списков прямоугольных треугольников."""
    M = int(input("Введите количество прямоугольных треугольников M: "))
    right_triangles_obj = [generate_random_right_triangle() for _ in range(M)]
    right_triangles_coords = [(triangle.A, triangle.B, triangle.C) for triangle in right_triangles_obj]
    return right_triangles_obj, right_triangles_coords


def compute_average_area(triangles):
    """Вычисление средней площади треугольников из списка."""
    # Вычисляем площадь каждого треугольника и находим среднее значение
    areas = [triangle.area() for triangle in triangles]
    return sum(areas) / len(areas)


def compute_min_perimeter(triangles):
    """Вычисление минимального периметра среди треугольников из списка."""
    # Вычисляем периметр каждого треугольника и находим минимальное значение
    perimeters = min([triangle.perimeter() for triangle in triangles])
    return (perimeters)


def find_triangle_with_max_hypotenuse(right_triangles):
    """Находим треугольник с наибольшей гипотенузой и выводим детали."""
    hypotenuses = [right_triangle.hypotenuse() for right_triangle in right_triangles]
    max_hypotenuse_value = max(hypotenuses)
    max_hypotenuse_index = hypotenuses.index(max_hypotenuse_value)
    triangle_with_max_hypotenuse = right_triangles[max_hypotenuse_index]
    return {
        "Номер треугольника": max_hypotenuse_index,
        "гипотенуза": max_hypotenuse_value,
        "треугольник с вершинами": str(triangle_with_max_hypotenuse)
    }




triangles_obj, triangles_coords = input_triangles()
right_triangles_obj, right_triangles_coords = input_right_triangles()


average_area = compute_average_area(triangles_obj)
min_perimeter = compute_min_perimeter(triangles_obj)
triangle_with_max_hypotenuse = find_triangle_with_max_hypotenuse(right_triangles_obj)





# Создание треугольника и его методы
triangle_test = Triangle(Point(1, 1), Point(2, 2), Point(0, 4))
print(f"Существование треугольника: {triangle_test.exists()}, длины сторон: {triangle_test.sides_length()}, углы(в градусах): {triangle_test.angles()}, периметр: {triangle_test.perimeter()}, площадь: {triangle_test.area()}")

# Создание прямоугольного треугольника и метод is_right
right_triangle_test = RightTriangle(Point(0, 0), Point(3, 0), Point(0, 4))
print(f"Существование прямоугольного треугольника: {right_triangle_test.exists()},\
 периметр: {right_triangle_test.perimeter()}, площадь: {right_triangle_test.area()}, прямоугольник треугольный: {right_triangle_test.is_right()}")

print(f"Список всех треугольников: {triangles_coords}")
print(f"Список всех прямоугольных треугольников: {right_triangles_coords}")

print(f"Средняя площадь всех треугольников: {average_area}")
print(f"Треугольник с минимальным периметром: {min_perimeter}")
print(f"Информация о треугольнике с наибольшей гипотенузой: {triangle_with_max_hypotenuse}")




def print_triangles_table_fstring(triangles):
    """Вывод данных о треугольниках в виде таблицы с использованием f-строк."""
    print("\n Таблица треугольников:")
    print("-" * 60)
    print(f"{'Треугольник':<10} | {'Точка A':<15} | {'Точка B':<15} | {'Точка C':<15}")
    print("-" * 60)
    for idx, triangle in enumerate(triangles, 1):
        print(f"{f'№{idx}':<10} | {str(triangle.A):<15} | {str(triangle.B):<15} | {str(triangle.C):<15}")
    print("\n")


def print_right_triangles_table_fstring(right_triangles):
    """Вывод данных о прямоугольных треугольниках в виде таблицы с использованием f-строк."""
    print("Таблица прямоугольных треугольников:")
    print("-" * 60)
    print(f"{'Треугольник':<15} | {'Точка A':<15} | {'Точка B':<15} | {'Точка C':<15}")
    print("-" * 60)
    for idx, right_triangle in enumerate(right_triangles, 1):
        print(f"{f'№{idx}':<15} | {str(right_triangle.A):<15} | {str(right_triangle.B):<15} | {str(right_triangle.C):<15}")
    print("\n")

print_triangles_table_fstring(triangles_obj)
print_right_triangles_table_fstring(right_triangles_obj)



def triangles_to_dataframe(triangles):
    """Преобразование списка объектов класса Triangle в DataFrame."""
    data = {
        "Точка A": [str(triangle.A) for triangle in triangles],
        "Точка B": [str(triangle.B) for triangle in triangles],
        "Точка C": [str(triangle.C) for triangle in triangles]
    }
    return pd.DataFrame(data)

def right_triangles_to_dataframe(right_triangles):
    """Преобразование списка объектов класса RightTriangle в DataFrame."""
    data = {
        "Точка A": [str(triangle.A) for triangle in right_triangles],
        "Точка B": [str(triangle.B) for triangle in right_triangles],
        "Точка C": [str(triangle.C) for triangle in right_triangles]
    }
    return pd.DataFrame(data)


triangles_df = triangles_to_dataframe(triangles_obj)
right_triangles_df = right_triangles_to_dataframe(right_triangles_obj)

print("Таблица треугольников:")
print(triangles_df)
print("\n Таблица прямоугольных треугольников:")
print(right_triangles_df)

```

Имеется 2 ящика с одним товаром и два мешка с другим. Выяснить, у какого продукта
соотношение масса/стоимость наименьшее.

```python
from abc import ABC, abstractmethod

class ProductInterface(ABC):
    """Интерфейс для продуктов"""
    @abstractmethod
    def mass_to_cost_ratio(self):
        """Метод для вычисления отношения массы к стоимости"""
        pass



class Box(ProductInterface):
    """Класс Ящик"""
    def __init__(self, name, unit_price, quantity, mass_per_unit, mass_of_product):
        """
        Инициализация ящика

        :param name: Наименование ящика
        :param unit_price: Цена единицы товара в ящике
        :param quantity: Количество единиц товара в ящике
        :param mass_per_unit: Масса одной единицы товара
        :param mass_of_product: Масса всего товара в ящике без учета упаковки
        """
        self.name = name
        self.unit_price = unit_price
        self.quantity = quantity
        self.mass_per_unit = mass_per_unit
        self.mass_of_product = mass_of_product


    @property
    def box_cost(self):
        """Вычисление стоимости ящика"""
        return self.unit_price * self.quantity


    @property
    def box_mass(self):
        """Вычисление массы ящика"""
        return self.mass_per_unit * self.quantity + self.mass_of_product


    def mass_to_cost_ratio(self):
        """Метод для вычисления отношения массы ящика к его стоимости"""
        return self.box_mass / self.box_cost



class Product(ProductInterface):
    """Класс Товар"""
    def __init__(self, name, mass, price):
        """
        Инициализация товара

        :param name: Наименование товара
        :param mass: Масса товара
        :param price: Цена товара
        """
        self.name = name
        self.mass = mass
        self.price = price


    def mass_to_cost_ratio(self):
        """Метод для вычисления отношения массы товара к его стоимости"""
        return self.mass / self.price



class Bag(ProductInterface):
    """Класс Мешок"""
    def __init__(self, name, mass, price_per_kg):
        """
        Инициализация мешка

        :param name: Наименование мешка
        :param mass: Масса мешка
        :param price_per_kg: Цена за килограмм содержимого мешка
        """
        self.name = name
        self.mass = mass
        self.price_per_kg = price_per_kg


    @property
    def bag_cost(self):
        """Вычисление стоимости мешка"""
        return self.price_per_kg * self.mass


    def mass_to_cost_ratio(self):
        """Метод для вычисления отношения массы мешка к его стоимости"""
        return self.mass / self.bag_cost




box1 = Box("Box1", 100, 5, 2, 10)
box2 = Box("Box2", 120, 4, 3, 8)

bag1 = Bag("Bag1", 10, 5)
bag2 = Bag("Bag2", 8, 6)

products = [box1, box2, bag1, bag2]

# Продукт с наименьшим соотношением масса/стоимость
min_ratio_product = min(products, key=lambda p: p.mass_to_cost_ratio())
min_ratio_product.name
```
