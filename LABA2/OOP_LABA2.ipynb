Дополните программы третьим классом С, имеющим три поля (x,y,z) с соответствующими геттерами и сеттерами. В окончательной версии программы создайте три объекта по одному для каждого класса, явным образом указав значения обязательно всех свойств.

![](21.png)

```python
from typing import List, Union


class A:
    """Основной класс A с одним свойством x."""
    def __init__(self, x=0):
        """Инициализация экземпляра класса A со свойством x."""
        self.__x = x

    @property
    def x(self):
        """Получение значения x."""
        return self.__x

    @x.setter
    def x(self, x):
        """Установка значения x."""
        self.__x = x



class B(A):
    """Класс B, наследующийся от класса A и вводящий новое свойство y."""

    def __init__(self, x=0, y=0):
        """Инициализация экземпляра класса B со свойствами x и y."""
        super().__init__(x)
        self.__y = y

    @property
    def y(self):
        """Получение значения y."""
        return self.__y

    @y.setter
    def y(self, y):
        """Установка значения y."""
        self.__y = y



class C(B):
    """Класс C, наследующийся от класса B и вводящий новое свойство z."""
    def __init__(self, x=0, y=0, z=0):
        """Инициализация экземпляра класса C со свойствами x, y и z."""
        super().__init__(x, y)
        self.__z = z

    @property
    def z(self):
        """Получение значения z."""
        return self.__z

    @z.setter
    def z(self, z):
        """Установка значения z."""
        self.__z = z



obj_a = A(10)
obj_b = B(10, 20)
obj_c = C(10, 20, 30)

obj_a_values = (obj_a.x)
obj_b_values = (obj_b.x, obj_b.y)
obj_c_values = (obj_c.x, obj_c.y, obj_c.z)

obj_a_values, obj_b_values, obj_c_values
```

### Задание 3.
Разберите, реализуйте в системе программирования и протестируйте программу, соответствующую примеру лекции: Пример 2 (слайд 9). Предусмотрите ввод данных круга и количество точек с клавиатуры, а координаты точек сгенерируйте случайным образом с последующим выводом данных на экран, используя соответствующий метод класса Tochka. Если в лекциях рассматривалась тема «Магические методы», то замените методы `vivod` на `__str__`.

![](23.png)

```python
from typing import List
import random

class Point:
    """ Класс, описывающий точку на плоскости. """
    def __init__(self, x=0, y=0):
        """ Инициализация точки с координатами x и y. """
        self.x = x
        self.y = y

    def __str__(self):
        """ Вывод координат точки в формате (x, y). """
        return f"({self.x}, {self.y})"

    def __repr__(self):
        """Представление точки для вывода в списках и других структурах. """
        return self.__str__()


class Circle(Point):
    """ Класс, описывающий круг на плоскости. """
    def __init__(self, x=0, y=0, r=0):
        """ Инициализация круга с координатами центра x, y и радиусом r. """
        super().__init__(x, y)
        self.r = r

    def distance(self, point):
        """ Расчет расстояния от центра круга до указанной точки. """
        return ((point.x - self.x) ** 2 + (point.y - self.y) ** 2) ** 0.5

    def __str__(self):
        """ Вывод информации о круге. """
        return f"Круг с центром в точке: ({self.x}, {self.y}) и радиусом: {self.r}"



def create_circle_object():
    """ Создание объекта Circle на основе пользовательского ввода. """
    x = int(input("Введите координату x: "))
    y = int(input("Введите координату y: "))
    r = int(input("Введите радиус r: "))
    return Circle(x, y, r)


def generate_random_coords(_) -> Point:
    """ Генерация случайных координат для точки. """
    x = random.randint(0, 20)
    y = random.randint(0, 20)
    return Point(x, y)


def generate_random_circle_coords() -> List[Point]:
    """Генерация списка из n точек со случайными координатами."""
    n = int(input("Введите количество точек: "))
    return list(map(generate_random_coords, range(n)))


def farthest_point_from_circle(circle, points_list):
    """Находит точку, которая наиболее удалена от центра указанного круга."""
    # Вычисляем расстояния от центра круга до каждой точки
    distances = [circle.distance(point) for point in points_list]
    # Находим индекс точки с наибольшим расстоянием
    max_index = distances.index(max(distances))
    return max_index, points_list[max_index]



circle = create_circle_object()
random_points = generate_random_circle_coords()

print(circle)
print(f"Все точки: {random_points}")
print(f"Точка расположенная дальше всех от центра круга и ее координаты: {farthest_point_from_circle(circle, random_points)}")
```
